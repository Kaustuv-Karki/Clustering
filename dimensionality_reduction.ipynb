```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
def perform_PCA(df):
    df = df.dropna()
    df_std = StandardScaler().fit_transform(df)
    pca_5 = PCA(n_components=5, random_state=20)
    pca_5.fit(df_std)
    reduced_dimension = pca_5.transform(df_std)
    return reduced_dimension
```

```python
reduced_def_features = perform_PCA(defence_feature)
```

```python
red_poss_features = perform_PCA(possession_featues)
```

```python
red_chance_features = perform_PCA(chance_creaton_features)
```

```python
red_passing_features = perform_PCA(passing_feature)
```

```python
red_attack_features = perform_PCA(attacking_features)
```
